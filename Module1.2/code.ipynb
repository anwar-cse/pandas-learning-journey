```python
import pandas as pd
data={
    "name":["Anwar","mominul","maruf","a","b","c","d"],
    "age":[23,12,45,2,3,4,5],
    "cgpa":[3.76,3.11,3.60,1,2,4,3]
}
df=pd.DataFrame(data)
print(df)
print("show 1st 5 rows")
print(df.head())
print("last 5 rows")
print(df.tail())
print("dataframe shape:")
print(df.shape)
print(df.columns)
print(df.dtypes)
#print(df.info())
print(df.describe())
```
