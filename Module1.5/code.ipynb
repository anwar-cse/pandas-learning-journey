Sorting

```python
import pandas as pd

data={
    "name":["anwar","rakib","habib","dhruva"],
    "age":[23,24,3,44],
    "result":[4,5,3,4]
}
df=pd.DataFrame(data)

df=df.sort_values(by=["age","result"])
print(df)
```

unique value

```python
print(df["result"].unique())
print(df["result"].nunique())
print(df["result"].value_counts())
print(df["result"].value_counts(normalize=True))
print(df["result"].value_counts().head(1))
```
