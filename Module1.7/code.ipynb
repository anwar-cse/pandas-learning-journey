create two dataframe

```python
import pandas as pd

df1=pd.DataFrame(
    {
        "id":[1,2,3],
        "name":["anwar","rakib","shuvo"],
        "age":[23,43,12]
    }
)

df2=pd.DataFrame(
    {
        "id":[1,2,4],
        "salary":[3000,2000,4000]
    }
)
```

Merge

```python
result=pd.merge(df1,df2,on="id")
print(result)
```

Merge Type

```python
#left join
result1=pd.merge(df1,df2,on="id",how="left")
print(result1)

#right join
result2=pd.merge(df1,df2,on="id",how="right")
print(result2)

#inner join
result3=pd.merge(df1,df2,on="id",how="inner")
print(result3)
```

Cocatenation

```python
result4=pd.concat([df1,df2])
print(result4)
result5=pd.concat([df1,df2],axis=1)
print(result5)
```

join

```python
result6=df1.join(df2,lsuffix="_left",rsuffix="_right")
print(result6)
```
