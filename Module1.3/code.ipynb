create dataframe

```python
import pandas as pd

data={
    "name":["anwar","korim","rohim","amjad"],
    "age":[23,43,12,34],
    "cgpa":[3.76,3.6,3.1,3.5]
}
df=pd.DataFrame(data)
```

data select

```python
print(df["name"])
print(df[["name","age"]])
print(df.loc[0])
print(df.loc[0:2])
print(df.iloc[1])
print(df.loc[0,"name"])
```

filtering

```python
print(df[df["age"]>23])
print(df[(df["name"] == "anwar") & (df["age"]>20)])
```
