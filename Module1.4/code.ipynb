create dataframe

```python
import pandas as pd

data={
    "name":["anwar",None,"anwar","adiyan","ojiha"],
    "age":[23,None,None,34,67],
    "cgpa":[2,3,2,1,2]
}

df=pd.DataFrame(data)
print(df)
```

data cleaning

```python
print(df.isnull())
print(df.isnull().sum())
#df=df.dropna()
#print(df)

df=df.dropna(axis=1)
print(df)
```

Null value fill

```python
import pandas as pd

data1={
    "name":["anwar",None,"anwar","adiyan","ojiha"],
    "age":[23,None,23,34,67],
    "cgpa":[2,3,2,1,2]
}

df=pd.DataFrame(data1)
print(df)

df["age"]=df["age"].fillna(0)
print(df)
```

rename

```python
df=df.rename(columns={"name":"Name"})
print(df)
```

duplicate value remove

```python
df=df.drop_duplicates()
print(df)
```

replace

```python
df["Name"]=df["Name"].replace("anwar","Anwar")
print(df)
```
